```python
import pandas as pd
import requests

# Initialize an empty list to store the tables
tables=[]
data = []
# Loop through pages 1 to 72
for i in range(1, 72):
    # Create the URL with the current page number
    url = f"https://stats.espncricinfo.com/ci/engine/stats/index.html?class=3;page=i;result=1;result=2;result=3;spanmax3=31+Dec+2100;spanmin3=01+Jan+2001;spanval3=span;template=results;type=team"
    headers = {'User-Agent': 'Mozilla/5.0'}
    # Make the request to the URL
    response = requests.get(url, headers=headers)

    #print the scrped data
    tables=pd.read_html(response.text)
    print(tables[2])
    data.append(tables[2])
    
```

```python
df = pd.concat(data, ignore_index=True)
df
```
